```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('bird.jpg', cv2.IMREAD_COLOR)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
kernel = np.ones((5,5)) / 25
blur_image = cv2.filter2D(image, -1, kernel)
plt.imshow(cv2.cvtColor(blur_image, cv2.COLOR_BGR2RGB))
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# laplacian kernel
kernel = np.array([[0, 1, 0],[1, -4, 1],[0, 1, 0]])
# cv2.Laplacian(gray_image, -1, kernel)

result_image = cv2.filter2D(gray_image, -1, kernel)

plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
```

```python
original_image = cv2.imread('unsharp_bird.jpg', cv2.IMREAD_COLOR)

# sharpen kernel
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

result_image = cv2.filter2D(original_image, -1, kernel)

plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
```
