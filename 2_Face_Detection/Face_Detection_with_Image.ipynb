```python
import cv2
import matplotlib.pyplot as plt 
```

```python
cascade_classifier = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
img = cv2.imread('girl_face.png')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
detected_img = cascade_classifier.detectMultiScale(image = gray_img, 
                                                   scaleFactor = 1.01, 
                                                   minNeighbors = 5, 
                                                   minSize = (30, 30))
'''
image        -> gri görüntü 
scaleFactor  -> her görüntü ölçeklemesinde arama penceresinin ne kadar küçüleceği
minNeighbors -> pozitip tespitlerin geçerli sayılabilmesi için gereken komşu tespit sayısı
minSize      -> Algılanacak en küçük nesne boyutu
'''

for (x, y, width, height) in detected_img:
    cv2.rectangle(img, (x, y), (x+width, y+height), (0, 0, 255), 10)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
