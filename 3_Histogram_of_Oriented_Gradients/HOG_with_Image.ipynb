```python
from skimage import data, feature
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
image = data.astronaut()
```

```python
image.shape
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
'''
orientation     -> histogram contains nine bins to angles [0, 20, 40, 60, 80, 100, 120, 140, 160] in degrees
pixel per cell  -> usually we combine 8x8 pixels and these are usually called 'patches'
cells per block -> for normalization (2, 2) 
block norm      -> L2
visualize       -> True , if False function returns just the hook features the one dimensional vector basically. 
'''
hog_vector , hog_image = feature.hog(image = image, 
                                    orientations = 9,
                                    pixels_per_cell = (8, 8), 
                                    cells_per_block = (2, 2), 
                                    block_norm = 'L2',
                                    visualize = True,
                                    channel_axis = -1)

# if image has color_channels value 3, channel_axis value is equal to -1
# if image has grayscale, channel_axis value is equal to None


fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 6))
axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(hog_image)
axes[1].set_title('HOG Image')
axes[1].axis("off")

plt.show()
```
